```python
# Create a year column based on student name
student_names = df_dom['Student'].value_counts().index.tolist()
# Move Arne G to be first element in list since he is 2015 student
student_names.insert(0, student_names.pop())

years = ['2015', '2016', '2017']

res = {student_names[i]: years[i] for i in range(len(student_names))}

df_dom['year'] = df_dom['Student'].replace(res)
df_dom.head()
```

```python
df_dom.drop(columns='Student', inplace=True)
```

```python
# We do not use 2015 lake data as there is not sufficient sampling information \
# and we do not know if samples were collected in the middle or at the shore of lakes
df_dom = df_dom.loc[~((df_dom['year'] == '2015') & (df_dom['System'] == 'lake')),:]
df_dom.head()
```

```python
# Split the data in 'Station' column into the site name which is the letter+number and data column
# Select only 2015 and 2016 as 2017 splits into 3, so do separate
df_dom_sub = df_dom.loc[(df_dom['year'] == '2015') | (df_dom['year'] == '2016')]
# There is a couple station strings that have -inc in them, split
df_dom_sub['Station'] = df_dom_sub['Station'].str.replace(r'-inc', '', regex=True)
df_dom_sub['Station'] = df_dom_sub['Station'].str.replace(r' – inc', '', regex=True)
df_dom_sub['Station'] = df_dom_sub['Station'].str.replace(r'/16', '', regex=True)

df_dom_sub['Station'].str.strip()

df_dom_sub['site'] = df_dom_sub.Station.str.split(' ', expand = True)[0]
df_dom_sub['date'] = df_dom_sub.Station.str.split(' ', expand = True)[1]
```

```python
# The station data for 2017 has an initial integer that we do not need
df_dom_17 = df_dom.loc[df_dom['year'] == '2017']
#df_dom_17['Station'].str.strip()
df_dom_17['site'] = df_dom_17.Station.str.split(' ', expand = True)[1]
df_dom_17['date'] = df_dom_17.Station.str.split(' ', expand = True)[2]
```

```python
df_dom = pd.concat([df_dom_sub, df_dom_17])
df_dom.drop(columns='Station', inplace=True)
```
